```python
import random
# 1 
x = range(60)
y_shanghai = [random.uniform(15, 18) for i in x]
# 
y_beijing = [random.uniform(1, 3) for i in x]
# 2 
figure, axes = plt.subplots(nrows=1, ncols=2, figsize=(20,8), dpi=80)
# 3 
axes[0].plot(x, y_shanghai, label='')
axes[1].plot(x, y_beijing, color='orange', linestyle='--', label='')
# 
axes[0].legend(loc='best')
axes[1].legend(loc='best')
# xy
x_label = ["11{}".format(i) for i in x]
axes[0].set_xticks(x[::5], x_label[::5])
axes[0].set_yticks(range(0, 40, 5))
axes[1].set_xticks(x[::5], x_label[::5])
axes[1].set_yticks(range(0, 40, 5))
# 
plt.rcParams['font.sans-serif'] = ['KaiTi'] 
plt.rcParams['axes.unicode_minus'] = False
# 
axes[0].grid(True, linestyle='--', alpha=0.5)
axes[1].grid(True, linestyle='--', alpha=0.5)
# 
axes[0].set_xlabel("")
axes[0].set_ylabel("")
axes[0].set_title("11012")
axes[1].set_xlabel("")
axes[1].set_ylabel("")
axes[1].set_title("11012")
# 4 
plt.show()
```
