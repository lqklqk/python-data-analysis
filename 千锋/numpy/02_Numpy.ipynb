# 一、创建ndarray

###### 1. 使用np.array()由python list创建

- 参数为列表： [1, 4, 2, 5, 3]

注意：

- numpy默认ndarray的所有元素的类型是相同的
- 如果传进来的列表中包含不同的类型，则统一为同一类型，优先级：str>float>int
- ndarray的数据类型：

  - int:   int8、uint8、int16、int32、int64
  - float:   float16、float32、float64
  - str:  字符串

```python
l = [1,4,2,3,5,6]
n = np.array(l)
n
```

```python
# 类型
type(n)
```

```python
# 形状
n.shape
```

```python
# 优先级：str>float>int
n2 = np.array([3.14,2,'hello'])
n2
```

###### 2. 使用np的常规函数创建

包含以下常见创建方法：

###### 1) np.ones(shape, dtype=None, order='C')

- 创建一个所有元素都为1的多维数组

参数说明:

- shape : 形状（元组）
- dtype=None: 元素类型

- order ： {‘C’，‘F’}，可选，默认值：C 是否在内存中以行主（C-风格）或列主（Fortran-风格）顺序存储多维数据, 一般默认即可

```python
# 创建元素为1的数组
n3 = np.ones(shape=(3,4), dtype=np.int16)
n3
```

###### 2) np.zeros(shape, dtype=float, order='C')

- 创建一个所有元素都为1的多维数组

参数说明:

- shape : 形状

- dtype=None: 元素类型

- order ： {‘C’，‘F’}，可选，默认值：C 是否在内存中以行主（C-风格）或列主（Fortran-风格）顺序存储多维数据, 一般默认即可

```python
# 创建一个所有元素都为0的多维数组
n4 = np.zeros(shape=(5,5), dtype=np.int16)
n4
```

###### 3) np.full(shape, fill_value, dtype=None, order='C')

- 创建一个所有元素都为指定元素的多维数组

参数说明:

- shape: 形状

- fill_value: 填充值

- dtype=None: 元素类型

- order ： {‘C’，‘F’}，可选，默认值：C 是否在内存中以行主（C-风格）或列主（Fortran-风格）顺序存储多维数据, 一般默认即可

```python
# 创建一个所有元素都为8的多维数组
n5 = np.full(shape=(2,3,4),fill_value=8,dtype=np.int8)
n5
```

###### 4) np.eye(N, M=None, k=0, dtype=float)

- 主对角线为1其他的位置为0的二维数组（单位矩阵）

参数说明:

- N: 行数

- M: 列数, 默认为None,表示和行数一样

- k=0: 向右偏移0个位置（主对角线偏移量，负数表示向左偏移）

- dtype=None: 元素类型

```python
n6 = np.eye(6, 6, dtype=int)
n6
```

```python
# 右偏移2个位置
n7 = np.eye(6,6, k=2, dtype=int)
n7
```

###### 5) np.linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None)

- 创建一个等差数列

参数说明:

- start: 开始值

- stop: 结束值

- num=50: 等差数列中默认有50个数，如果包含头尾数字，所以一般设置为51，元素全部设置为整数

- endpoint=True: 是否包含结束值

- retstep=False: 是否返回等差值(步长)

- dtype=None: 元素类型

```python
n8 = np.linspace(0, 100,num=51,dtype=np.int8)
n8
```

```python
n9 = np.linspace(0, 100, num=50, endpoint=False)
n9
```

```python
# 显示差值（步长）: retstep=True
n10 = np.linspace(0, 100, num=51,retstep=True)
n10
```

###### 6) np.arange([start, ]stop, [step, ]dtype=None)

- 创建一个数值范围的数组
- 和Python中range功能类似

参数说明:

- start : 开始值(可选)

- stop: 结束值(不包含)

- step: 步长(可选)

- dtype=None: 元素类型

```python
n11 = np.arange(2,10,3)
n11
```

###### 7) np.random.randint(low, high=None, size=None, dtype='l')

- 创建一个随机整数的多维数组

参数说明:

- low : 最小值

- high=None: 最大值

  - **high=None时，生成的数值在[0, low)区间内**
  - **如果使用high这个值，则生成的数值在[low, high)区间**

- size=None: 数组形状, 默认只输出一个随机值

- dtype=None: 元素类型b

```python
# 随机整数（生成一个数字）
n12 = np.random.randint(10)
display(n12, type(n12))
```

```python
# 随机取多维数组: 一维
n13 = np.random.randint(0, 10, size=6)
display(n13, type(n13))
```

```python
# 随机取多维数组: 三维
n14 = np.random.randint(0, 256, size=(20,40,3))
display(n14, type(n14))
```

```python
plt.imshow(n14)  # 显示图片
```

###### 9)np.random.normal(loc=0.0, scale=1.0, size=None)

- 创建一个服从**正态分布**的多维数组

参数说明:

- loc=0.0: 均值, 对应着正态分布的中心

- scale: 标准差, 对应分布的宽度，scale越大，正态分布的曲线越矮胖，scale越小，曲线越高瘦

- size=None: 数组形状

```python
# 正态分布
n18 = np.random.normal(loc=100, scale=5, size=(3,4))
n18
```

- 创建一个元素为0~1(左闭右开)的随机数的多维数组

参数说明:

- size=None: 数组形状

```python
# 生成0~1的随机数
n19 = np.random.random(size=(3,4))
n19
```

###### 11) np.random.rand(d0, d1, ..., dn)

- 创建一个元素为0~1(左闭右开)的随机数的多维数组
- 和np.random.random功能类似, 掌握其中一个即可

参数说明:

- dn : 第n个维度的数值

```python
# 生成0~1的随机数
n20 = np.random.rand(4,5)
n20
```

# 三、ndarray的基本操作

## 1. 索引
### 一维与列表完全一致, 多维时同理

```python
l2 = [1,2,3,4,5,6]
l2[0],l2[-1]
```

```python
nd1 = np.array(l2)
nd1[0],nd1[-1]
```

```python
# 二维数组
nd2 = np.random.randint(0,10,size=(4,5))
nd2
```

```python
# 找到最后一个数
nd2[-1][-1]
```

```python
# 简写
nd2[-1,-1]
```

```python
# 三维数组
nd3 = np.random.randint(0,100,size=(4,5,6))
nd3
```

```python
# 找到
nd3[1,2,-1]
```

### 根据索引修改数据

```python
# 修改某个元素
nd3[1,2,-1] = 999
nd3[1,2,-1]
```

```python
# 修改某个一维数组（或者二维数组）
nd3[0][1] = [99]
```
