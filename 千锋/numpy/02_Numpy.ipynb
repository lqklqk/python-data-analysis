###### 2. 使用np的常规函数创建

包含以下常见创建方法：

###### 1) np.ones(shape, dtype=None, order='C')

- 创建一个所有元素都为1的多维数组

参数说明:

- shape : 形状（元组）
- dtype=None: 元素类型

- order ： {‘C’，‘F’}，可选，默认值：C 是否在内存中以行主（C-风格）或列主（Fortran-风格）顺序存储多维数据, 一般默认即可

```python
# 创建元素为1的数组
n3 = np.ones(shape=(3,4), dtype=np.int16)
n3
```

###### 2) np.zeros(shape, dtype=float, order='C')

- 创建一个所有元素都为1的多维数组

参数说明:

- shape : 形状

- dtype=None: 元素类型

- order ： {‘C’，‘F’}，可选，默认值：C 是否在内存中以行主（C-风格）或列主（Fortran-风格）顺序存储多维数据, 一般默认即可

```python
# 创建一个所有元素都为0的多维数组
n4 = np.zeros(shape=(5,5), dtype=np.int16)
n4
```

###### 3) np.full(shape, fill_value, dtype=None, order='C')

- 创建一个所有元素都为指定元素的多维数组

参数说明:

- shape: 形状

- fill_value: 填充值

- dtype=None: 元素类型

- order ： {‘C’，‘F’}，可选，默认值：C 是否在内存中以行主（C-风格）或列主（Fortran-风格）顺序存储多维数据, 一般默认即可

```python
# 创建一个所有元素都为8的多维数组
n5 = np.full(shape=(2,3,4),fill_value=8,dtype=np.int8)
n5
```

###### 4) np.eye(N, M=None, k=0, dtype=float)

- 主对角线为1其他的位置为0的二维数组（单位矩阵）

参数说明:

- N: 行数

- M: 列数, 默认为None,表示和行数一样

- k=0: 向右偏移0个位置（主对角线偏移量，负数表示向左偏移）

- dtype=None: 元素类型

```python
n6 = np.eye(6, 6, dtype=int)
n6
```

```python
# 右偏移2个位置
n7 = np.eye(6,6, k=2, dtype=int)
n7
```

###### 5) np.linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None)

- 创建一个等差数列

参数说明:

- start: 开始值

- stop: 结束值

- num=50: 等差数列中默认有50个数，如果包含头尾数字，所以一般设置为51，元素全部设置为整数

- endpoint=True: 是否包含结束值

- retstep=False: 是否返回等差值(步长)

- dtype=None: 元素类型

```python
n8 = np.linspace(0, 100,num=51,dtype=np.int8)
n8
```

```python
n9 = np.linspace(0, 100, num=50, endpoint=False)
n9
```

```python
# 显示差值（步长）: retstep=True
n10 = np.linspace(0, 100, num=51,retstep=True)
n10
```

## 2. 切片
### 一维与列表完全一致 多维时同理

```python
# python 列表的切片
l3 = [1,2,3,4,5,6,7,8,9]
l3[2:6], l3[::-1]
```

## 3. 变形

### 使用reshape函数, 注意参数是一个tuple
- reshape函数会生成新的数组，原数组形状不变
### 使用resize
- resize函数不会生成新数组，会修改原数组形状
### 变形之后的数据总数与原来的保持一致

```python
nd7 = np.arange(1, 21)
nd7
```

```python
# reshape 将 ndarray 形状做改变
nd7.shape
```

```python
# 变成二维
nd8 = np.reshape(nd7,shape=(4,5))
nd8
```

```python
nd9 = nd8.reshape((5,4))
nd9
```

```python
# 变成一维
nd9.reshape((20))
nd9.reshape((-1))
# 使用-1，表示任意剩余维度长度
nd9.reshape((4,-1))
nd7.reshape(-1,5)
nd9.reshape((-1,2))
# 变成3维
nd7.reshape(2,5,2)
```

```python
# 使用 resize 函数
nd7.resize((4,5))
nd7
```

#### 类型修改
- astype函数，将数组中的元素转为指定类型，返回新的数组，原有数组元素类型不变
- tostring函数、tobytes函数，将数组中的元素转为字节类型
  - 注意，转成bytes后没有dtype属性
  - tostring函数已经被废弃，不建议使用

```python
nd10=nd7.astype(str)
nd10.dtype,nd7.dtype
```

```python
nd11=nd7.tobytes()
nd11,nd7.dtype
```

```python
nd12=nd7.tostring()
nd12,nd7.dtype
```

#### 数组去重
- set函数（集合，不存在重复元素，默认去重），该函数只能接收一维数组，可以结合ndarry.flatten()函数使用，返回的是集合而不是ndarray
- ndarry.flatten()，将多维数组转为一维数组
- np.unique(需要去重的数组)函数，去重之后返回新的一维数组（去重之后的数组返回的是一维数组）

```python
nd13 = np.random.randint(0,10,size=(3,6,2))
nd13
```

```python
nd14 = nd13.flatten()
np.unique(nd13), set(nd14)
```

## 4. 级联

### np.concatenate()
- 参数是列表或元组
- 级联的数组维度必须相同
  - 如果是行合并（默认，axis=0），那么列数必须相同
  - 如果是列合并（axis=1），那么行数必须相同
- 可通过axis参数改变级联的方向
### np.hstack 与 np.vstack
- np.hstack是列合并（水平合并），行数必须一致，相当于np.concatenate((n1,n2),axis=1)
- np.vstack是列合并（垂直合并），列数必须相同，相当于np.concatenate((n1,n2),axis=0)

```python
n11 = np.random.randint(0,10,size=(3,6))
n12 = np.random.randint(0,10,(3,7))
n11,n12
```

```python
np.concatenate((n11,n12),axis=1)
```
